```python
SHEET_ID = '1Yjmg552R22Z-GYifTWBlHpmbSQPQYXGz_0mdJ-0WJvQ'
SHEET_NAME = 'data_entry'
url = f'https://docs.google.com/spreadsheets/d/{SHEET_ID}/gviz/tq?tqx=out:csv&sheet={SHEET_NAME}'
data = pd.read_csv(url)
```

```python
n_participants = int(data.columns[-1])
```

```python
data = data.iloc[:,:6]
data.columns = ['Observer','Sl','Group_Symbol','Round','Player_ID','Score']
condition = (data['Observer'].isna()) #& (data['Player_ID'].isna())
#data = data[~data['Observer'].isna()]
data = data[~condition]
data.shape 
try:
    master_data.shape
except NameError:
    master_data = pd.DataFrame()
master_data = pd.concat([master_data,data],ignore_index=True)
```

```python
round_name = data['Round'].unique()[0]
prev_round_name = 'Round '+ str(int(round_name.split(' ')[1])-1)
try:
    groups_dict
except NameError:
    groups_dict ={}
dump = pd.DataFrame()
team_names = ['Star','Square','Triangle','Circle','Hexagon','Rectangle']
for item in team_names:
    ids = list(master_data[master_data['Group_Symbol']==item]['Player_ID'])
    ids = [int(x) for x in ids]
    dump[item] = ids
if round_name=='Round 1':
    groups_dict[prev_round_name] = dump
else:
    pass
del dump
```

```python
master_data = master_data.reset_index()
master_data = master_data.drop_duplicates()
group_scores = master_data.groupby('Group_Symbol')['Score'].sum()
master_data_pivot = master_data.pivot(index='Player_ID', columns='Round', values='Score')
```

```python
master_data_pivot = pd.read_csv('master_data_pivot.csv')
```

```python
rank_df = pd.DataFrame(master_data_pivot.sum(axis=1),columns=['score'])
rank_df['rank']=rank_df['score'].rank(ascending=False)
rank_df = rank_df.sort_values('rank')
rank_df = rank_df.reset_index()
```

```python
rank_df.columns = ['Player_ID','score','rank']
```

```python
rank_df['Player_ID'] = rank_df['Player_ID'].astype(int)
rank_list = list(rank_df['Player_ID'])
groups = [rank_list[i:i+10] for i in range(0,len(rank_df),10)]
groups = np.array(groups)
groups = pd.DataFrame(groups)
groups = groups.T
groups.columns = team_names
```

```python
groups_dict[round_name] = groups
groups_dict = dict(sorted(groups_dict.items()))
```
